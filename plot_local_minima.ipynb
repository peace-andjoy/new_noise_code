```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def function(x):
    return 0.2 * x**4 - 2 * x**2 + 4 * np.sin(x) + 8

# 生成 x 值的范围
x = np.linspace(-5, 5, 400)

# 计算函数的值
y = function(x)

y[0] = 40

# 绘制函数图像
plt.plot(x, y, label='Objective Function')

# 绘制局部平坦最小值
plt.scatter([-1.8, 0, 2.5], [function(-1.8), function(0), function(2.5)],
            color='red', marker='o', label='Local Flat Minima')

# 绘制尖锐最小值
plt.scatter([1.8], [function(1.8)], color='green', marker='^', label='Sharp Minima')

# 添加图例和标签
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Local Flat Minima vs. Sharp Minima')

# 显示图形
plt.show()
```
