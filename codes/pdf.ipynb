```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, levy_stable, t
plt.style.use('default')

# 
alpha = 1  # alpha-stablealpha
beta = 0.0  # alpha-stablebeta
truncation_min = -0.5
truncation_max = 0.5
sample_size = 10000

# truncated alpha-stable
alpha_stable_samples = levy_stable.rvs(alpha, beta, scale=0.028, size=sample_size)

# 
alpha_stable_samples[alpha_stable_samples < truncation_min] = truncation_min
alpha_stable_samples[alpha_stable_samples > truncation_max] = truncation_max

# # truncated Gaussian
# gaussian_samples = norm.rvs(loc=0, scale=0.1, size=sample_size)

# # 
# gaussian_samples[gaussian_samples < truncation_min] = truncation_min
# gaussian_samples[gaussian_samples > truncation_max] = truncation_max

# truncated t
t_samples = t.rvs(1.5, scale=0.03, size=sample_size)

# # 
# gaussian_samples[gaussian_samples < truncation_min] = truncation_min
# gaussian_samples[gaussian_samples > truncation_max] = truncation_max

# 
t_samples[t_samples < truncation_min] = truncation_min
t_samples[t_samples > truncation_max] = truncation_max

# 
plt.figure(dpi=300)
plt.hist(alpha_stable_samples, bins=50, alpha=0.5, density=True, label='Clipped $\\alpha$-Stable')
# plt.hist(gaussian_samples, bins=50, alpha=0.5, density=True, label='Clipped Gaussian')
plt.hist(t_samples, bins=50, alpha=0.5, density=True, label='Clipped t')

# 
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable

#  alpha 
alphas = [2, 1.5, 1, 0.5]

#  x 
x = np.linspace(-5, 5, 1000)

#  alpha-stable  PDF 
plt.figure(dpi=300)
style = ['solid', 'dashed', 'dotted', 'dashdot']
for i, alpha in enumerate(alphas):
    plt.plot(x, levy_stable.pdf(x, alpha, 0), label=f'$\\alpha$ = {alpha}', linestyle=style[i])

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
