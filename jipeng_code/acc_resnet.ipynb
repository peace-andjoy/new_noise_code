```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pickle
import os.path
```

```python
network = 'resnet'
sigma = 0.02
```

```python
summary_acc = pd.DataFrame()
# acc
for alpha in [0, 2, 1.9, 1.5, 1.3, 1, 0.9]:

    acc_path = "test_accuracy_{}.txt".format(alpha)
    if os.path.exists(acc_path) == 0:
        print('no exists, alpha={}'.format(alpha))
    else:
#         f = open(acc_path, 'rb')
#         acc = read(acc_path)
#         df = pd.DataFrame(acc)
        acc = np.loadtxt(acc_path).reshape(1,7)
        df = pd.DataFrame(acc)
        col = 'clean' if alpha ==0 else alpha
        tmp = pd.DataFrame(df.mean().round(decimals=4).astype(str) + ' (std=' + df.std().round(decimals=4).astype(str) + ')', columns = [col])
        summary_acc = pd.concat([summary_acc, tmp], axis=1)
# summary_acc.index = ['clean',2,1.9,1.5,1.3,1,0.9]
# summary_acc = summary_acc.T
```

```python
summary_acc.index = ['clean',2,1.9,1.5,1.3,1,0.9]
summary_acc = summary_acc.T
summary_acc
```
